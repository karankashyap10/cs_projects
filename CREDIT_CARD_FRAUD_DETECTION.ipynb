IMPORTING THE DEPENDENCIES


```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Load the dataset into the dataframe
df = pd.read_csv('/content/creditcard[1].csv')
```

```python
#First five rows
df.head()
```

```python
#Dataset information
df.info()
```

```python
#Checking the number of null values
df.isnull().sum()
```

```python
#Distribution of legit and fraudlent transactions
df['Class'].value_counts()
```

This Dataset is highly unbalanced.
0 -> Legit transactions
1 -> Fraudlent Transactions

```python
#Separating the Data for analysis
Legit = df[df.Class == 0]
Fraud = df[df.Class == 1]
```

```python
print(Legit.shape)
print(Fraud.shape)
```

```python
#Statistical measures of the data
Legit.describe()
```

```python
Fraud.describe()
```

UnderSampling


Build a sample dataset containing similar distribution of normal transactions and Fraudlent transactions

Number of Fraudlent Transactions -> 492

```python
Legit_sample = Legit.sample(n = 492)
```

```python
#Concatenating two dataframes, row-wise
new_dataset = pd.concat([Legit_sample,Fraud],axis=0)
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into Features and Targets

```python
X = new_dataset.drop(columns = 'Class',axis=1)
Y = new_dataset['Class']
```

Splitting the data into training and testing data.

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify = Y, random_state=2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

MODEL TRAINING

```python
model = LogisticRegression(max_iter=200,solver='liblinear')
```

```python
#training the model with training data
model.fit(X_train,Y_train)
```

MODEL EVALUATION

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
print('Accuracy on training data is : - ',training_data_accuracy)
```

```python
#accuracy score on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
print('Accuracy on test data is : - ',test_data_accuracy)
```

Saving the Trained Model

```python
import pickle
```
